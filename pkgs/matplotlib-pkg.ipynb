###### Matplotlib visualisation tool for ploting graphs-scatterplots-lineplots-3Dplots-etc with the most popupar pyplot module

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
```

```python
# PLOTING A SIMPLE LINE PLOT GRAPH
plt.plot(x, y)
```

```python
# PLOTING A SIMPLE SCATTER PLOT GRAPH
plt.scatter(x[::10], y[::10], color="#000000")
```

```python
# THIS PLOT THE LINES IN THE SAME GRAPH
plt.plot(x, y, color="r")
plt.plot(x, np.cos(y), color="g")
```

```python
# THIS PLOT THE LINE IN A DASH LINE FORMAT
plt.plot(x, y, "--c")
```

```python
# THIS PLOT THE LINE IN A DASHDOT LINE FORMAT
plt.plot(x, y, "-.c")
```

```python
# THIS PLOT THE LINE IN A DOTTED LINE FORMAT
plt.plot(x, y, ":c")
```
